# PROBLEM STATEMENT  :  Whether the person has survived the incident or not ?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
titanic_data=pd.read_csv('train.csv')
```

```python
titanic_data.head()
```

* the Survived column implies whether the passenger had survived or not
* the pclass column implies about the type of ticket he had bought.(1st class means that the most expensive one, and so on)
* sibsp column about the number of siblings/spouses they had onboard the ship
* parch means the number of parents onboard the ship
* Embarked means that at which port, he/she had boarded the ship

# Data Preprocessing

### Number of rows and columns

```python
titanic_data.shape
```

```python
titanic_data.info()
```

### Null values Dealing


```python
titanic_data.isnull().sum()
```

### Drop the 'Cabin' column

```python
titanic_data=titanic_data.drop(columns='Cabin',axis=1)
```

### Replacing the missing 'age' values with the mean age.

```python
titanic_data['Age'].fillna(titanic_data['Age'].mean(),inplace=True)
```

```python
titanic_data.isnull().sum()
```

### Also, replacing the missing Embarked values with the 'mode' values of the 'Embarked' column

```python
titanic_data['Embarked'].mode()[0]
```

```python
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0],inplace=True)
```

```python
titanic_data.head()
```

### Now, there will be no missing value(s) for any column

```python
titanic_data.isnull().sum()
```

# Encoding the categorical columns

```python
titanic_data['Embarked'].value_counts()
```

#### Converting the categorical columns 

```python
titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_data.head()
```

```python
titanic_data.tail()
```

# Separating features and target

```python
X = titanic_data.drop(columns = ['PassengerId','Name','Ticket','Survived'],axis=1)
Y = titanic_data['Survived']
```

```python
X
```

```python
Y
```

# Generally, train data is just 80% of total data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

# Model Training

# Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

# Model Evaluation

## Evaluation on Training Data

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data : ', training_data_accuracy)
```

## Evaluation on Test Data

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of test data : ', test_data_accuracy)
```
