# Import necessary libraries

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('athlete_events.csv')
region_df=pd.read_csv('noc_regions.csv')
```

```python
df.tail()
```

## We only want the summer Olympics Season

```python
df.shape
```

```python
df=df[df['Season']=='Summer']
```

```python
df.shape
```

```python
region_df.head()
```

### merging the two datas.

```python
df=df.merge(region_df,on='NOC',how='left')
```

## Total number of different unique regions

```python
df['region'].unique().shape[0]
```

```python
df.isnull().sum()
```

### Drop the duplicates

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

## Categorizing the medals in basis of gold,silver or bronze

```python
df.head()
```

```python
pd.get_dummies(df['Medal'])
```

```python
df=pd.concat([df,pd.get_dummies(df['Medal'])],axis=1)
```

```python
df.head()
```

### So wherever medals are 'NaN', either of gold, silver or bronze are 0. 

```python
df.groupby('NOC').sum()
```

```python
df.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
```

### However,this data is NOT right. THIS is because, like suppose if india had won a hockey match once in 1990, then it is equivalent to 11 hockey medals here. since a team consist of 11 members, so 11 medals by that logic. hence, thats why, we want to remove this anamoly.

### We can ensure this by manually just dropping the duplicates [ Teams, NOC, Year, Season, City, Sport, Event, Medal ]

```python
df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])
```

```python
medal_tally=df.drop_duplicates(subset=['Team','NOC','Games','Year','City','Sport','Event','Medal'])
```

```python
medal_tally=medal_tally.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold',ascending=False).reset_index()
```

```python
medal_tally['Total']=medal_tally['Gold']+medal_tally['Silver']+medal_tally['Bronze']
```
