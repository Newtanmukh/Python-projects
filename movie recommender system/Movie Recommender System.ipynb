### Convert Function

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):  #ast library will do exactly this. it will convert the string to the list of dictonaries.
        L.append(i['name']) 
    return L 
```

```python

#This could have been used, unfortunately thr genres thing is encoded inside a string and is NOT A LIST.
def get_tags(x):
    lists=[]
    for y in x:
           lists.append(y["name"])
    return lists

movies['genre']=movies['genres'].apply(lambda x:convert(x))
```

### Get only the first three Cast Names

```python
def get_first_three_cast_names(text):
    L = []
    for index,i in enumerate(ast.literal_eval(text)):  #ast library will do exactly this. it will convert the string to the list of dictonaries.
        if index==3:
            break
        L.append(i['name']) 
    return L 

```

```python
movies['cast']=movies['cast'].apply(lambda x:get_first_three_cast_names(x))
```

```python
movies.head()
```

### Get only the director from the crew

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):# use ast again,coz those in string and we want the list inside of it.
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 

```

```python
movies['crew']=movies['crew'].apply(lambda x:fetch_director(x))
```

## Remove the empty space and combine them

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

## NOTE: But we have a problem here, words like 'love', 'loving', 'lover' etc. are very similar so it will be a big problem.

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i)) # ps.stem(i) will basically convert loving,lover,loved etc to love.
    return " ".join(y)
```

```python
new['tags']=new['tags'].apply(lambda x:stem(x))
```
