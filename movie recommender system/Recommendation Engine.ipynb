## Convert genres and keywords to just the names inside.

```python
import ast
#we need to import ast, since the genres is a string and the list of the dictonaries is inside that.

def convert(obj):
    list=[]
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```
